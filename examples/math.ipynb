```python
import math
import random
from itertools import permutations
```

```python
def generate_list():
    _list = []
    for i in range(9):
        for j in range(i, 9):
            for k in range(j, 9):
                for l in range(k, 9):
                    for m in range(l ,9):
                        _list.append([i+1, j+1, k+1, l+1, m+1])
    return _list
```

```python
result = generate_list()
result
len(result)
```
